### Introdução
####	Breve descrição da base de dados

O banco de dados utilizado é referente a mudanças de temperatura, coletado e disponibilizado pelo FAOSTAT (Food and Agriculture Organization of the United Nations) na mudança média de temperatura por país. O banco de dados cobre o período de 1961-2020. Os dados estão disponíveis por mês, estação e ano (considerando anomalias médias anuais, i.e., mudanças de temperatura com respeito a uma climatologia de base, correspondendo ao período de 1951-1980).

Fonte: https://www.kaggle.com/datasets/sevgisarac/temperature-change/data

O intuito dessa análise foi comparar as mudanças relativas de temperatura em 3 subsets de países, escolhidos de acordo com critérios geográficos e populacionais.

Com base no artigo "Change in cooling degree days with global mean temperature rise increasing from 1.5 °C to 2.0 °C", publicado na revista Nature, embora países na África central e subsaariana vem sofrendo e sofrerão com os maiores aumentos em temperatura em termos absolutos, são países no norte da Europa que sofrem e sofrerão com os maiores aumentos em temperatura em termos relativos.

A partir da análise do banco de dados da FAOSTAT, criamos gráficos que facilitam a visualização destes dados e corroboram o que foi apontado pelo estudo.

Fonte: https://www.nature.com/articles/s41893-023-01155-z
Fonte: https://theconversation.com/northern-europe-faces-biggest-relative-increase-in-uncomfortable-heat-and-is-dangerously-unprepared-new-research-209745

### Processamento

Para o tratamento de dados foram criados subsets baseados em regiões geográficas e características populacionais, de modo a facilitar a visualização e processamento de dados.

Foram removidas colunas julgadas desnecessárias e dados nulos. Também foram normalizados os valores de temperatura para plotagem do gráfico.

Para os gráficos, foram escolhidos 1 país de cada subset e os valores correspondentes ao verão dos respectivos hemisférios.

### Conclusão

Os gráficos permitem a fácil visualização de um grave problema, possivelmente o maior desafio que a humanidade já enfrentou.

O resultado é consistente com o que têm apresentado inúmeras organizações como o IPCC (Intergovernmental Panel on Climate Change) e a NASA (National Aeronautics and Space Administration), que alertam para os perigos da mudança climática impulsionada pela queima de combustíveis fósseis e o efeito estufa.

Embora os gráficos explicitem uma mudança relativa em temperaturas do norte da Europa, não devemos achar que o problema não será grave também nos trópicos. Uma análise futura poderá fazer uso de dados absolutos e não relativos para o propósito de alerta.

A análise cruzada com outros dados, como aqueles relativos à emissão de CO², podem nos dar uma ideia das desigualdades globais de um problema compartilhado. Fica o alerta de que os países do Sul Global sofrerão os devastadores efeitos de um problema que, historicamente, não é de sua responsabilidade.

Fonte: https://www.carbonbrief.org/analysis-which-countries-are-historically-responsible-for-climate-change/

### Manipulando os dados: Criação de subsets e remoção de valores

```python
# Criação do subset 1:
# 5 países mais populosos da América do Sul 
# e seus respectivos códigos para este DF:
#   Brazil (21), Colombia (44), Argentina (9), Peru (170), Venezuela (236)

# selecionando apenas as linhas correspondetes a estes 5 países

subset1 = df[df['Area Code (FAO)'].isin([21,44,9,170,236])]

subset1
```

```python
# Criação do subset 2:
# 5 países europeus apontados no relatório
# como particularmente afetados pelo aumento de temperatura em termos relativos
# e seus respectivos códigos para este DF:
#   Switzerland (211), United Kingdom (229), Norway (162), Finland (67), Ireland (104)

# selecionando apenas as linhas correspondentes a estes 5 países

subset2 = df[df['Area Code (FAO)'].isin([211,229,162,67,104])]

subset2
```

```python
# Criação do subset 3:
# 5 países africanos apontados no relatório 
# como particularmente afetados pelo aumento de temperatura em termos absolutos
# e seus respectivos códigos para este DF:
#   Central African Republic (37), Burkina Faso (233), Mali (133), Sudan* (206), Nigeria (159)

# * As medições referentes ao antigo Sudão foram escolhidas como parâmetro
# em detrimento de Sudão e Sudão do Sul considerando o processo de separação recente do país

# # selecionando apenas as linhas correspondetes a estes 5 países

subset3 = df[df['Area Code (FAO)'].isin([37,233,133,206,159])]
```

```python
# eliminando as seguintes colunas:
## 'Domain Code' (1 valor: ET)
## 'Domain' (1 valor: Temperature change)
## 'Element Code' (1 valor: 7271 (Temperature change))
## 'Element' (1 valor: Temperature change)
## 'Unit' (1 valor: ºC)
## 'Flag'
## 'Flag Description' (incidência de 3% de 'Data not avaiable')

# na função drop especificar parâmetro axis: {0 or ‘index’, 1 or ‘columns’}, default 0
  #  Whether to drop labels from the index (0 or ‘index’) or columns (1 or ‘columns’).

subset1 = subset1.drop(['Unit', 'Domain Code', 'Domain', 'Element Code', 'Element', 'Unit', 'Flag', 'Flag Description'], axis=1)
```

```python
subset2 = subset2.drop(['Unit', 'Domain Code', 'Domain', 'Element Code', 'Element', 'Unit', 'Flag', 'Flag Description'], axis=1)
```

```python
subset3 = subset3.drop(['Unit', 'Domain Code', 'Domain', 'Element Code', 'Element', 'Unit', 'Flag', 'Flag Description'], axis=1)
```

```python
# verificando se há dados nulos no subset 1 (América do Sul)

subset1.isnull().sum()
```

```python
# verificando se há nulos no subset 2 (Europa)

subset2.isnull().sum()
```

```python
# verificando se há nulos no subset 3 (África)

subset3.isnull().sum()
```

```python
# removendo dados nulos 

subset3 = subset3.dropna(subset=['Value'])

# verificando se a remoção foi feita
subset3.isnull().sum()
```

### Analisando os dados

```python
# Normalização de colunas numéricas subset 1

subset1_norm = subset1.copy()

for column in subset1_norm.columns:
    subset1_norm["Value"] = (subset1_norm["Value"] - subset1_norm["Value"].min()) / (subset1_norm["Value"].max() - subset1_norm["Value"].min())
    
subset1_norm
```

```python
# Normalização de colunas numéricas subset 2

subset2_norm = subset2.copy()

for column in subset2_norm.columns:
    subset2_norm["Value"] = (subset2_norm["Value"] - subset2_norm["Value"].min()) / (subset2_norm["Value"].max() - subset2_norm["Value"].min())
    
subset2_norm
```

```python
# Normalização de colunas numéricas subset 3

subset3_norm = subset3.copy()

for column in subset3_norm.columns:
    subset3_norm["Value"] = (subset3_norm["Value"] - subset3_norm["Value"].min()) / (subset3_norm["Value"].max() - subset3_norm["Value"].min())
    
subset3_norm
```

```python
#Extract datas
country='Brazil'
month='Dec?Jan?Feb'
temps=[]
years=[]

for index, row in subset1.iterrows():
    if row["Months"]==month and row["Area"]==country:
        temps.append(row["Value"])
        years.append(row["Year"])
        
print(temps)
print(years)
```

```python
# plotando os dados

fig, ax = plt.subplots()

# Create a colormap (from blue to red) based on temperature values
cmap = plt.get_cmap('coolwarm')  

# Create the bar chart with the colormap
bars = ax.bar(years, temps, color=cmap(temps))

ax.bar(years, temps)
ax.set_title('Temperature change - base 1961–2020 - {} - {}'.format(country, month))
ax.set_ylabel('Δ°C')
bars = ax.bar(years, temps, color=cmap(temps))
plt.savefig('temp_change_braz.png')
```
